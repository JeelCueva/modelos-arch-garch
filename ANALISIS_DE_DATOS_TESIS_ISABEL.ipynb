# PARTE 3: ESTAD√çSTICAS DESCRIPTIVAS

```python
# ============================================================================
# PARTE 3: ESTAD√çSTICAS DESCRIPTIVAS
# ============================================================================

print("üìà AN√ÅLISIS ESTAD√çSTICO DESCRIPTIVO")
print("="*60)

# Estad√≠sticas b√°sicas para Calidad de Servicio
print("\nüìä CALIDAD DE SERVICIO (n=70) - ESTAD√çSTICAS GENERALES:")
print("-"*50)
calidad_stats = df_calidad.iloc[:, 1:].describe()
print(calidad_stats.round(2))

# Estad√≠sticas b√°sicas para Procesos Administrativos
print("\nüîß PROCESOS ADMINISTRATIVOS (n=18) - ESTAD√çSTICAS GENERALES:")
print("-"*50)
procesos_stats = df_procesos.iloc[:, 1:].describe()
print(procesos_stats.round(2))

# Crear dimensiones para an√°lisis
print("\nüîç CREANDO DIMENSIONES PARA AN√ÅLISIS...")

# Dimensiones Calidad de Servicio
df_calidad['DIM_CONFIABILIDAD'] = df_calidad[['CONF_1', 'CONF_2', 'CONF_3', 'CONF_4']].mean(axis=1)
df_calidad['DIM_RESPONSABILIDAD'] = df_calidad[['RESP_5', 'RESP_6', 'RESP_7', 'RESP_8']].mean(axis=1)
df_calidad['DIM_SEGURIDAD'] = df_calidad[['SEG_9', 'SEG_10', 'SEG_11', 'SEG_12']].mean(axis=1)
df_calidad['DIM_EMPATIA'] = df_calidad[['EMP_13', 'EMP_14', 'EMP_15', 'EMP_16']].mean(axis=1)
df_calidad['DIM_TANGIBLES'] = df_calidad[['TANG_17', 'TANG_18', 'TANG_19']].mean(axis=1)
df_calidad['DIM_SERVICIOS_TI'] = df_calidad[['TI_20', 'TI_21', 'TI_22', 'TI_23']].mean(axis=1)

# Variable global Calidad
dimensiones_calidad = ['DIM_CONFIABILIDAD', 'DIM_RESPONSABILIDAD', 'DIM_SEGURIDAD',
                      'DIM_EMPATIA', 'DIM_TANGIBLES', 'DIM_SERVICIOS_TI']
df_calidad['CALIDAD_GLOBAL'] = df_calidad[dimensiones_calidad].mean(axis=1)

# Dimensiones Procesos Administrativos
df_procesos['DIM_PLANIFICACION'] = df_procesos[['PLAN_1', 'PLAN_2', 'PLAN_3']].mean(axis=1)
df_procesos['DIM_ORGANIZACION'] = df_procesos[['ORG_4', 'ORG_5', 'ORG_6']].mean(axis=1)
df_procesos['DIM_DIRECCION'] = df_procesos[['DIR_7', 'DIR_8', 'DIR_9']].mean(axis=1)
df_procesos['DIM_CONTROL'] = df_procesos[['CTRL_10', 'CTRL_11', 'CTRL_12']].mean(axis=1)
df_procesos['DIM_INTEGRACION_TI'] = df_procesos[['TI_ADMIN_13', 'TI_ADMIN_14', 'TI_ADMIN_15']].mean(axis=1)

# Variable global Procesos
dimensiones_procesos = ['DIM_PLANIFICACION', 'DIM_ORGANIZACION', 'DIM_DIRECCION',
                       'DIM_CONTROL', 'DIM_INTEGRACION_TI']
df_procesos['PROCESOS_GLOBAL'] = df_procesos[dimensiones_procesos].mean(axis=1)

print("‚úÖ Dimensiones creadas exitosamente")

# An√°lisis por dimensiones - Calidad de Servicio
print("\nüìä AN√ÅLISIS POR DIMENSIONES - CALIDAD DE SERVICIO:")
print("-"*50)
for i, dim in enumerate(dimensiones_calidad, 1):
    media = df_calidad[dim].mean()
    std = df_calidad[dim].std()
    minimo = df_calidad[dim].min()
    maximo = df_calidad[dim].max()
    print(f"{i}. {dim.replace('DIM_', '').ljust(15)}: M={media:.2f} ¬± {std:.2f} (Rango: {minimo:.1f}-{maximo:.1f})")

print(f"\nüéØ CALIDAD GLOBAL: {df_calidad['CALIDAD_GLOBAL'].mean():.2f} ¬± {df_calidad['CALIDAD_GLOBAL'].std():.2f}")

# An√°lisis por dimensiones - Procesos Administrativos
print("\nüîß AN√ÅLISIS POR DIMENSIONES - PROCESOS ADMINISTRATIVOS:")
print("-"*50)
for i, dim in enumerate(dimensiones_procesos, 1):
    media = df_procesos[dim].mean()
    std = df_procesos[dim].std()
    minimo = df_procesos[dim].min()
    maximo = df_procesos[dim].max()
    print(f"{i}. {dim.replace('DIM_', '').ljust(15)}: M={media:.2f} ¬± {std:.2f} (Rango: {minimo:.1f}-{maximo:.1f})")

print(f"\nüéØ PROCESOS GLOBAL: {df_procesos['PROCESOS_GLOBAL'].mean():.2f} ¬± {df_procesos['PROCESOS_GLOBAL'].std():.2f}")

# Identificar dimensiones con mejor y peor puntuaci√≥n
print("\nüèÜ RANKING DE DIMENSIONES:")
print("-"*50)

# Para Calidad
medias_calidad = [(dim.replace('DIM_', ''), df_calidad[dim].mean()) for dim in dimensiones_calidad]
medias_calidad.sort(key=lambda x: x[1], reverse=True)

print("üìä CALIDAD DE SERVICIO (de mayor a menor puntuaci√≥n):")
for i, (nombre, media) in enumerate(medias_calidad, 1):
    estado = "üü¢ Fortaleza" if media > 3.5 else "üü° Promedio" if media > 3.0 else "üî¥ √Årea cr√≠tica"
    print(f"   {i}. {nombre.ljust(15)}: {media:.2f} {estado}")

# Para Procesos
medias_procesos = [(dim.replace('DIM_', ''), df_procesos[dim].mean()) for dim in dimensiones_procesos]
medias_procesos.sort(key=lambda x: x[1], reverse=True)

print("\nüîß PROCESOS ADMINISTRATIVOS (de mayor a menor puntuaci√≥n):")
for i, (nombre, media) in enumerate(medias_procesos, 1):
    estado = "üü¢ Fortaleza" if media > 3.5 else "üü° Promedio" if media > 3.0 else "üî¥ √Årea cr√≠tica"
    print(f"   {i}. {nombre.ljust(15)}: {media:.2f} {estado}")

# Distribuci√≥n de frecuencias
print("\nüìä DISTRIBUCI√ìN DE RESPUESTAS:")
print("-"*50)

def analizar_distribucion(df, nombre, cols_analizar):
    print(f"\n{nombre}:")
    data_flat = df[cols_analizar].values.flatten()
    data_flat = data_flat[~pd.isna(data_flat)]  # Remover NaN si los hay

    conteos = pd.Series(data_flat).value_counts().sort_index()
    total = len(data_flat)

    for valor in range(1, 6):
        count = conteos.get(valor, 0)
        porcentaje = (count / total) * 100
        barra = "‚ñà" * int(porcentaje / 2)
        print(f"   Valor {valor}: {count:3d} ({porcentaje:5.1f}%) {barra}")

    print(f"   Media: {np.mean(data_flat):.2f}")
    print(f"   Mediana: {np.median(data_flat):.2f}")

# Analizar distribuciones
analizar_distribucion(df_calidad, "CALIDAD DE SERVICIO",
                     [col for col in df_calidad.columns if col.startswith(('CONF_', 'RESP_', 'SEG_', 'EMP_', 'TANG_', 'TI_'))])

analizar_distribucion
```

# PARTE 4: AN√ÅLISIS DE CONFIABILIDAD (ALPHA DE CRONBACH)

```python
# ============================================================================
# PARTE 4: AN√ÅLISIS DE CONFIABILIDAD (ALPHA DE CRONBACH)
# ============================================================================

print("üîç AN√ÅLISIS DE CONFIABILIDAD - ALPHA DE CRONBACH")
print("="*60)

def cronbach_alpha(df_items):
    """
    Calcula el Alpha de Cronbach para medir la consistencia interna
    """
    # Convertir a matriz numpy
    items = df_items.values
    n_items = items.shape[1]

    # Calcular varianzas
    item_variances = np.var(items, axis=0, ddof=1)
    total_variance = np.var(np.sum(items, axis=1), ddof=1)

    # Formula del Alpha de Cronbach
    alpha = (n_items / (n_items - 1)) * (1 - np.sum(item_variances) / total_variance)

    return alpha

def interpretar_alpha(alpha):
    """Interpreta el valor del Alpha de Cronbach"""
    if alpha >= 0.9:
        return "Excelente"
    elif alpha >= 0.8:
        return "Bueno"
    elif alpha >= 0.7:
        return "Aceptable"
    elif alpha >= 0.6:
        return "Cuestionable"
    else:
        return "Pobre"

print("\nüìä CONFIABILIDAD POR DIMENSIONES - CALIDAD DE SERVICIO:")
print("-"*55)

# Confiabilidad Calidad de Servicio
dimensiones_calidad_vars = {
    'CONFIABILIDAD': ['CONF_1', 'CONF_2', 'CONF_3', 'CONF_4'],
    'RESPONSABILIDAD': ['RESP_5', 'RESP_6', 'RESP_7', 'RESP_8'],
    'SEGURIDAD': ['SEG_9', 'SEG_10', 'SEG_11', 'SEG_12'],
    'EMPAT√çA': ['EMP_13', 'EMP_14', 'EMP_15', 'EMP_16'],
    'TANGIBLES': ['TANG_17', 'TANG_18', 'TANG_19'],
    'SERVICIOS_TI': ['TI_20', 'TI_21', 'TI_22', 'TI_23']
}

alphas_calidad = {}
for dimension, variables in dimensiones_calidad_vars.items():
    alpha = cronbach_alpha(df_calidad[variables])
    interpretacion = interpretar_alpha(alpha)
    alphas_calidad[dimension] = alpha

    emoji = "üü¢" if alpha >= 0.7 else "üü°" if alpha >= 0.6 else "üî¥"
    print(f"   {emoji} {dimension.ljust(15)}: Œ± = {alpha:.3f} ({interpretacion})")

# Alpha total para Calidad de Servicio
todas_vars_calidad = [var for vars_list in dimensiones_calidad_vars.values() for var in vars_list]
alpha_total_calidad = cronbach_alpha(df_calidad[todas_vars_calidad])
print(f"\nüéØ CALIDAD TOTAL (23 √≠tems): Œ± = {alpha_total_calidad:.3f} ({interpretar_alpha(alpha_total_calidad)})")

print("\nüîß CONFIABILIDAD POR DIMENSIONES - PROCESOS ADMINISTRATIVOS:")
print("-"*55)

# Confiabilidad Procesos Administrativos
dimensiones_procesos_vars = {
    'PLANIFICACI√ìN': ['PLAN_1', 'PLAN_2', 'PLAN_3'],
    'ORGANIZACI√ìN': ['ORG_4', 'ORG_5', 'ORG_6'],
    'DIRECCI√ìN': ['DIR_7', 'DIR_8', 'DIR_9'],
    'CONTROL': ['CTRL_10', 'CTRL_11', 'CTRL_12'],
    'INTEGRACI√ìN_TI': ['TI_ADMIN_13', 'TI_ADMIN_14', 'TI_ADMIN_15']
}

alphas_procesos = {}
for dimension, variables in dimensiones_procesos_vars.items():
    alpha = cronbach_alpha(df_procesos[variables])
    interpretacion = interpretar_alpha(alpha)
    alphas_procesos[dimension] = alpha

    emoji = "üü¢" if alpha >= 0.7 else "üü°" if alpha >= 0.6 else "üî¥"
    print(f"   {emoji} {dimension.ljust(15)}: Œ± = {alpha:.3f} ({interpretacion})")

# Alpha total para Procesos Administrativos
todas_vars_procesos = [var for vars_list in dimensiones_procesos_vars.values() for var in vars_list]
alpha_total_procesos = cronbach_alpha(df_procesos[todas_vars_procesos])
print(f"\nüéØ PROCESOS TOTAL (15 √≠tems): Œ± = {alpha_total_procesos:.3f} ({interpretar_alpha(alpha_total_procesos)})")

# An√°lisis detallado de confiabilidad
print("\nüìã RESUMEN DE CONFIABILIDAD:")
print("-"*50)

print("üìä CALIDAD DE SERVICIO:")
dimensiones_problematicas_calidad = []
dimensiones_buenas_calidad = []

for dim, alpha in alphas_calidad.items():
    if alpha < 0.7:
        dimensiones_problematicas_calidad.append((dim, alpha))
    else:
        dimensiones_buenas_calidad.append((dim, alpha))

if dimensiones_buenas_calidad:
    print("   ‚úÖ Dimensiones con buena confiabilidad:")
    for dim, alpha in dimensiones_buenas_calidad:
        print(f"      ‚Ä¢ {dim}: Œ± = {alpha:.3f}")

if dimensiones_problematicas_calidad:
    print("   ‚ö†Ô∏è Dimensiones que requieren revisi√≥n:")
    for dim, alpha in dimensiones_problematicas_calidad:
        print(f"      ‚Ä¢ {dim}: Œ± = {alpha:.3f}")

print("\nüîß PROCESOS ADMINISTRATIVOS:")
dimensiones_problematicas_procesos = []
dimensiones_buenas_procesos = []

for dim, alpha in alphas_procesos.items():
    if alpha < 0.7:
        dimensiones_problematicas_procesos.append((dim, alpha))
    else:
        dimensiones_buenas_procesos.append((dim, alpha))

if dimensiones_buenas_procesos:
    print("   ‚úÖ Dimensiones con buena confiabilidad:")
    for dim, alpha in dimensiones_buenas_procesos:
        print(f"      ‚Ä¢ {dim}: Œ± = {alpha:.3f}")

if dimensiones_problematicas_procesos:
    print("   ‚ö†Ô∏è Dimensiones que requieren revisi√≥n:")
    for dim, alpha in dimensiones_problematicas_procesos:
        print(f"      ‚Ä¢ {dim}: Œ± = {alpha:.3f}")

# Crear DataFrame con resultados de confiabilidad
confiabilidad_results = pd.DataFrame({
    'Variable': ['Calidad Total'] + list(alphas_calidad.keys()) + ['Procesos Total'] + list(alphas_procesos.keys()),
    'Alpha_Cronbach': [alpha_total_calidad] + list(alphas_calidad.values()) + [alpha_total_procesos] + list(alphas_procesos.values()),
    'Interpretacion': [interpretar_alpha(alpha_total_calidad)] + [interpretar_alpha(a) for a in alphas_calidad.values()] +
                     [interpretar_alpha(alpha_total_procesos)] + [interpretar_alpha(a) for a in alphas_procesos.values()],
    'Tipo': ['Calidad'] * (len(alphas_calidad) + 1) + ['Procesos'] * (len(alphas_procesos) + 1)
})

print("\nüìä TABLA RESUMEN DE CONFIABILIDAD:")
print(confiabilidad_results.round(3))

# Recomendaciones basadas en confiabilidad
print("\nüí° RECOMENDACIONES BASADAS EN CONFIABILIDAD:")
print("-"*50)

if alpha_total_calidad >= 0.8 and alpha_total_procesos >= 0.8:
    print("‚úÖ Ambos instrumentos muestran buena confiabilidad general")
elif alpha_total_calidad >= 0.7 and alpha_total_procesos >= 0.7:
    print("‚úÖ Ambos instrumentos muestran confiabilidad aceptable")
else:
    print("‚ö†Ô∏è Uno o ambos instrumentos podr√≠an beneficiarse de revisi√≥n")

if len(dimensiones_problematicas_calidad) > 0 or len(dimensiones_problematicas_procesos) > 0:
    print("üîç Se sugiere revisar las dimensiones con Œ± < 0.7")
    print("üìù Considerar reformular √≠tems o agregar m√°s preguntas")

print("\n‚úÖ AN√ÅLISIS DE CONFIABILIDAD COMPLETADO")
print("üìå SIGUIENTE PASO: Ejecutar Parte 5 - An√°lisis de correlaciones")
```

```python
# ============================================================================
# PARTE 5: AN√ÅLISIS DE CORRELACIONES
# ============================================================================

print("üîó AN√ÅLISIS DE CORRELACIONES")
print("="*60)

# Funci√≥n para interpretar correlaciones
def interpretar_correlacion(r):
    """Interpreta el coeficiente de correlaci√≥n de Pearson"""
    r_abs = abs(r)
    if r_abs >= 0.8:
        return "Muy fuerte"
    elif r_abs >= 0.6:
        return "Fuerte"
    elif r_abs >= 0.4:
        return "Moderada"
    elif r_abs >= 0.2:
        return "D√©bil"
    else:
        return "Muy d√©bil"

# Configurar visualizaci√≥n
plt.figure(figsize=(15, 12))

# CORRELACIONES ENTRE DIMENSIONES DE CALIDAD DE SERVICIO
print("\nüìä CORRELACIONES ENTRE DIMENSIONES - CALIDAD DE SERVICIO:")
print("-"*55)

corr_calidad = df_calidad[dimensiones_calidad].corr()
print(corr_calidad.round(3))

# Visualizar matriz de correlaci√≥n - Calidad
plt.subplot(2, 2, 1)
sns.heatmap(corr_calidad, annot=True, cmap='RdYlBu_r', center=0,
            square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Correlaciones - Calidad de Servicio', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# CORRELACIONES ENTRE DIMENSIONES DE PROCESOS ADMINISTRATIVOS
print("\nüîß CORRELACIONES ENTRE DIMENSIONES - PROCESOS ADMINISTRATIVOS:")
print("-"*55)

corr_procesos = df_procesos[dimensiones_procesos].corr()
print(corr_procesos.round(3))

# Visualizar matriz de correlaci√≥n - Procesos
plt.subplot(2, 2, 2)
sns.heatmap(corr_procesos, annot=True, cmap='RdYlBu_r', center=0,
            square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Correlaciones - Procesos Administrativos', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# AN√ÅLISIS DE CORRELACIONES ESPEC√çFICAS
print("\nüîç AN√ÅLISIS DETALLADO DE CORRELACIONES:")
print("-"*50)

# Correlaciones m√°s altas en Calidad
print("üìä CALIDAD DE SERVICIO - Correlaciones m√°s fuertes:")
correlaciones_calidad = []
for i in range(len(dimensiones_calidad)):
    for j in range(i+1, len(dimensiones_calidad)):
        dim1 = dimensiones_calidad[i]
        dim2 = dimensiones_calidad[j]
        r = corr_calidad.loc[dim1, dim2]
        correlaciones_calidad.append((dim1.replace('DIM_', ''), dim2.replace('DIM_', ''), r))

correlaciones_calidad.sort(key=lambda x: abs(x[2]), reverse=True)

for i, (dim1, dim2, r) in enumerate(correlaciones_calidad[:5], 1):
    interpretacion = interpretar_correlacion(r)
    print(f"   {i}. {dim1} ‚Üî {dim2}: r = {r:.3f} ({interpretacion})")

# Correlaciones m√°s altas en Procesos
print("\nüîß PROCESOS ADMINISTRATIVOS - Correlaciones m√°s fuertes:")
correlaciones_procesos = []
for i in range(len(dimensiones_procesos)):
    for j in range(i+1, len(dimensiones_procesos)):
        dim1 = dimensiones_procesos[i]
        dim2 = dimensiones_procesos[j]
        r = corr_procesos.loc[dim1, dim2]
        correlaciones_procesos.append((dim1.replace('DIM_', ''), dim2.replace('DIM_', ''), r))

correlaciones_procesos.sort(key=lambda x: abs(x[2]), reverse=True)

for i, (dim1, dim2, r) in enumerate(correlaciones_procesos[:5], 1):
    interpretacion = interpretar_correlacion(r)
    print(f"   {i}. {dim1} ‚Üî {dim2}: r = {r:.3f} ({interpretacion})")

# CORRELACI√ìN PRINCIPAL: PROCESOS ‚Üî CALIDAD
print("\nüéØ AN√ÅLISIS DE RELACI√ìN PRINCIPAL: PROCESOS ‚Üí CALIDAD")
print("="*60)

# Para correlacionar, usamos los primeros 18 casos de calidad (mismo n que procesos)
df_calidad_18 = df_calidad.head(18).copy()

# Correlaci√≥n entre variables globales
correlacion_principal = df_calidad_18['CALIDAD_GLOBAL'].corr(df_procesos['PROCESOS_GLOBAL'])
interpretacion_principal = interpretar_correlacion(correlacion_principal)

print(f"üìà CORRELACI√ìN PROCESOS GLOBAL ‚Üî CALIDAD GLOBAL:")
print(f"   r = {correlacion_principal:.3f}")
print(f"   Interpretaci√≥n: {interpretacion_principal}")

# Test de significancia
from scipy.stats import pearsonr
stat, p_value = pearsonr(df_calidad_18['CALIDAD_GLOBAL'], df_procesos['PROCESOS_GLOBAL'])

print(f"   p-valor: {p_value:.3f}")
if p_value < 0.05:
    print(f"   ‚úÖ Correlaci√≥n estad√≠sticamente significativa (p < 0.05)")
else:
    print(f"   ‚ö†Ô∏è Correlaci√≥n no estad√≠sticamente significativa (p ‚â• 0.05)")

# Visualizaci√≥n de la correlaci√≥n principal
plt.subplot(2, 2, 3)
plt.scatter(df_procesos['PROCESOS_GLOBAL'], df_calidad_18['CALIDAD_GLOBAL'],
           alpha=0.7, s=60, color='steelblue', edgecolors='black', linewidth=0.5)

# L√≠nea de tendencia
z = np.polyfit(df_procesos['PROCESOS_GLOBAL'], df_calidad_18['CALIDAD_GLOBAL'], 1)
p = np.poly1d(z)
plt.plot(df_procesos['PROCESOS_GLOBAL'], p(df_procesos['PROCESOS_GLOBAL']),
         "r--", alpha=0.8, linewidth=2)

plt.xlabel('Procesos Administrativos (Global)')
plt.ylabel('Calidad de Servicio (Global)')
plt.title(f'Correlaci√≥n Principal\nr = {correlacion_principal:.3f}',
          fontsize=12, fontweight='bold')
plt.grid(True, alpha=0.3)

# CORRELACIONES CRUZADAS ENTRE DIMENSIONES
print(f"\nüîÑ CORRELACIONES CRUZADAS (PROCESOS ‚Üí CALIDAD):")
print("-"*50)

correlaciones_cruzadas = []
for dim_proc in dimensiones_procesos:
    for dim_cal in dimensiones_calidad:
        r = df_procesos[dim_proc].corr(df_calidad_18[dim_cal])
        correlaciones_cruzadas.append((
            dim_proc.replace('DIM_', ''),
            dim_cal.replace('DIM_', ''),
            r
        ))

# Ordenar por valor absoluto
correlaciones_cruzadas.sort(key=lambda x: abs(x[2]), reverse=True)

print("üèÜ TOP 10 CORRELACIONES CRUZADAS:")
for i, (proc, cal, r) in enumerate(correlaciones_cruzadas[:10], 1):
    interpretacion = interpretar_correlacion(r)
    print(f"   {i:2d}. {proc.ljust(12)} ‚Üí {cal.ljust(12)}: r = {r:6.3f} ({interpretacion})")

# Crear matriz de correlaciones cruzadas para visualizaci√≥n
matriz_cruzada = np.zeros((len(dimensiones_procesos), len(dimensiones_calidad)))
for i, dim_proc in enumerate(dimensiones_procesos):
    for j, dim_cal in enumerate(dimensiones_calidad):
        matriz_cruzada[i, j] = df_procesos[dim_proc].corr(df_calidad_18[dim_cal])

# Visualizar correlaciones cruzadas
plt.subplot(2, 2, 4)
sns.heatmap(matriz_cruzada,
           xticklabels=[d.replace('DIM_', '') for d in dimensiones_calidad],
           yticklabels=[d.replace('DIM_', '') for d in dimensiones_procesos],
           annot=True, cmap='RdYlBu_r', center=0,
           fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Correlaciones Cruzadas\n(Procesos ‚Üí Calidad)', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()

# RESUMEN DE CORRELACIONES
print(f"\nüìã RESUMEN DE AN√ÅLISIS DE CORRELACIONES:")
print("="*50)

print(f"üéØ Correlaci√≥n principal (Procesos ‚Üí Calidad): {correlacion_principal:.3f} ({interpretacion_principal})")
print(f"üìä Significancia estad√≠stica: {'‚úÖ S√≠' if p_value < 0.05 else '‚ùå No'} (p = {p_value:.3f})")

# Identificar las correlaciones m√°s prometedoras
correlaciones_fuertes = [c for c in correlaciones_cruzadas if abs(c[2]) >= 0.5]
if correlaciones_fuertes:
    print(f"\nüî• RELACIONES M√ÅS PROMETEDORAS (|r| ‚â• 0.5):")
    for proc, cal, r in correlaciones_fuertes[:5]:
        print(f"   ‚Ä¢ {proc} ‚Üí {cal}: r = {r:.3f}")
else:
    print(f"\nüí° No se encontraron correlaciones muy fuertes (|r| ‚â• 0.5)")
    correlaciones_moderadas = [c for c in correlaciones_cruzadas if abs(c[2]) >= 0.3]
    if correlaciones_moderadas:
        print(f"üîç RELACIONES MODERADAS (|r| ‚â• 0.3):")
        for proc, cal, r in correlaciones_moderadas[:5]:
            print(f"   ‚Ä¢ {proc} ‚Üí {cal}: r = {r:.3f}")

print("\n‚úÖ AN√ÅLISIS DE CORRELACIONES COMPLETADO")
print("üìå SIGUIENTE PASO: Ejecutar Parte 6 - An√°lisis de regresi√≥n")
```

# PARTE 6: AN√ÅLISIS DE REGRESI√ìN Y PRUEBAS DE HIP√ìTESIS

```python
# ============================================================================
# PARTE 6: AN√ÅLISIS DE REGRESI√ìN Y PRUEBAS DE HIP√ìTESIS
# ============================================================================

print("üìà AN√ÅLISIS DE REGRESI√ìN Y PRUEBAS DE HIP√ìTESIS")
print("="*60)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from scipy.stats import f_oneway, normaltest

# REGRESI√ìN LINEAL SIMPLE: PROCESOS ‚Üí CALIDAD
print("\nüéØ REGRESI√ìN LINEAL: PROCESOS ADMINISTRATIVOS ‚Üí CALIDAD DE SERVICIO")
print("-"*65)

# Preparar datos (usando n=18 pareado)
X = df_procesos['PROCESOS_GLOBAL'].values.reshape(-1, 1)
y = df_calidad_18['CALIDAD_GLOBAL'].values

# Ajustar modelo
modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

# M√©tricas del modelo
r2 = r2_score(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r_pearson = np.corrcoef(X.flatten(), y)[0, 1]

print(f"üìä ECUACI√ìN DE REGRESI√ìN:")
print(f"   Calidad = {modelo.intercept_:.3f} + {modelo.coef_[0]:.3f} √ó Procesos")
print(f"")
print(f"üìà M√âTRICAS DEL MODELO:")
print(f"   R¬≤ = {r2:.3f} (Varianza explicada: {r2*100:.1f}%)")
print(f"   Correlaci√≥n (r) = {r_pearson:.3f}")
print(f"   RMSE = {rmse:.3f}")
print(f"   n = {len(X)} casos")

# Interpretaci√≥n del modelo
print(f"\nüí° INTERPRETACI√ìN:")
if modelo.coef_[0] > 0:
    print(f"   ‚úÖ Relaci√≥n POSITIVA: Por cada punto de mejora en procesos,")
    print(f"      la calidad mejora en {modelo.coef_[0]:.3f} puntos")
else:
    print(f"   ‚ö†Ô∏è Relaci√≥n NEGATIVA: Por cada punto de mejora en procesos,")
    print(f"      la calidad disminuye en {abs(modelo.coef_[0]):.3f} puntos")

if r2 >= 0.5:
    print(f"   üéØ MODELO FUERTE: Explica {r2*100:.1f}% de la variabilidad")
elif r2 >= 0.3:
    print(f"   üéØ MODELO MODERADO: Explica {r2*100:.1f}% de la variabilidad")
else:
    print(f"   üéØ MODELO D√âBIL: Solo explica {r2*100:.1f}% de la variabilidad")

# Visualizaci√≥n de la regresi√≥n
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.scatter(X, y, alpha=0.7, s=60, color='steelblue', edgecolors='black', linewidth=0.5)
plt.plot(X, y_pred, color='red', linewidth=2, label=f'y = {modelo.intercept_:.2f} + {modelo.coef_[0]:.2f}x')
plt.xlabel('Procesos Administrativos (Global)')
plt.ylabel('Calidad de Servicio (Global)')
plt.title(f'Regresi√≥n Lineal\nR¬≤ = {r2:.3f}', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

# An√°lisis de residuos
residuos = y - y_pred

plt.subplot(2, 2, 2)
plt.scatter(y_pred, residuos, alpha=0.7, s=60, color='orange', edgecolors='black', linewidth=0.5)
plt.axhline(y=0, color='red', linestyle='--', alpha=0.8)
plt.xlabel('Valores Predichos')
plt.ylabel('Residuos')
plt.title('Gr√°fico de Residuos', fontweight='bold')
plt.grid(True, alpha=0.3)

# Test de normalidad de residuos
stat_norm, p_norm = normaltest(residuos)
print(f"\nüîç DIAGN√ìSTICOS DEL MODELO:")
print(f"   Test de normalidad de residuos: p = {p_norm:.3f}")
if p_norm > 0.05:
    print(f"   ‚úÖ Residuos siguen distribuci√≥n normal (p > 0.05)")
else:
    print(f"   ‚ö†Ô∏è Residuos no siguen distribuci√≥n normal (p ‚â§ 0.05)")

# PRUEBA DE HIP√ìTESIS PRINCIPAL
print(f"\nüéØ PRUEBA DE HIP√ìTESIS PRINCIPAL")
print("="*50)

print(f"H‚ÇÄ: Los procesos administrativos NO influyen en la calidad del servicio (œÅ = 0)")
print(f"H‚ÇÅ: Los procesos administrativos S√ç influyen en la calidad del servicio (œÅ ‚â† 0)")
print(f"")
print(f"üìä RESULTADOS:")

# Test de correlaci√≥n
stat_corr, p_corr = pearsonr(X.flatten(), y)
print(f"   Correlaci√≥n observada: r = {stat_corr:.3f}")
print(f"   p-valor: {p_corr:.3f}")
print(f"   Nivel de significancia: Œ± = 0.05")
print(f"")

if p_corr < 0.05:
    print(f"‚úÖ DECISI√ìN: Se RECHAZA H‚ÇÄ")
    print(f"   Existe evidencia estad√≠stica significativa de que los procesos")
    print(f"   administrativos influyen en la calidad del servicio (p < 0.05)")
    conclusion = "SIGNIFICATIVA"
else:
    print(f"‚ùå DECISI√ìN: NO se puede rechazar H‚ÇÄ")
    print(f"   No hay evidencia estad√≠stica suficiente de relaci√≥n")
    print(f"   significativa entre las variables (p ‚â• 0.05)")
    conclusion = "NO SIGNIFICATIVA"

# REGRESI√ìN M√öLTIPLE POR DIMENSIONES
print(f"\nüìä AN√ÅLISIS DE REGRESI√ìN M√öLTIPLE")
print("-"*50)

# Preparar datos para regresi√≥n m√∫ltiple
X_multiple = df_procesos[dimensiones_procesos].values
y_multiple = df_calidad_18['CALIDAD_GLOBAL'].values

# Ajustar modelo m√∫ltiple
modelo_multiple = LinearRegression()
modelo_multiple.fit(X_multiple, y_multiple)
y_pred_multiple = modelo_multiple.predict(X_multiple)

# M√©tricas del modelo m√∫ltiple
r2_multiple = r2_score(y_multiple, y_pred_multiple)
rmse_multiple = np.sqrt(mean_squared_error(y_multiple, y_pred_multiple))

print(f"üìà MODELO DE REGRESI√ìN M√öLTIPLE:")
print(f"   R¬≤ = {r2_multiple:.3f} (Varianza explicada: {r2_multiple*100:.1f}%)")
print(f"   RMSE = {rmse_multiple:.3f}")
print(f"")
print(f"üîç COEFICIENTES POR DIMENSI√ìN:")

for i, dimension in enumerate(dimensiones_procesos):
    coef = modelo_multiple.coef_[i]
    dim_name = dimension.replace('DIM_', '')
    impacto = "Positivo" if coef > 0 else "Negativo"
    print(f"   {dim_name.ljust(15)}: Œ≤ = {coef:6.3f} ({impacto})")

print(f"   Intercepto: {modelo_multiple.intercept_:.3f}")

# Comparaci√≥n de modelos
print(f"\nüìä COMPARACI√ìN DE MODELOS:")
print("-"*35)
print(f"   Modelo Simple:   R¬≤ = {r2:.3f}")
print(f"   Modelo M√∫ltiple: R¬≤ = {r2_multiple:.3f}")

if r2_multiple > r2:
    mejora = r2_multiple - r2
    print(f"   ‚úÖ El modelo m√∫ltiple explica {mejora*100:.1f}% m√°s variabilidad")
else:
    print(f"   ‚ö†Ô∏è El modelo simple es igual o mejor")

# AN√ÅLISIS POR GRUPOS
print(f"\nüìä AN√ÅLISIS COMPARATIVO POR GRUPOS")
print("-"*50)

# Crear grupos basados en la mediana
mediana_procesos = df_procesos['PROCESOS_GLOBAL'].median()
mediana_calidad = df_calidad['CALIDAD_GLOBAL'].median()

# Grupos para Procesos (toda la muestra n=18)
df_procesos['GRUPO_PROCESOS'] = df_procesos['PROCESOS_GLOBAL'].apply(
    lambda x: 'Altos' if x > mediana_procesos else 'Bajos'
)

# Grupos para Calidad (toda la muestra n=70)
df_calidad['GRUPO_CALIDAD'] = df_calidad['CALIDAD_GLOBAL'].apply(
    lambda x: 'Alta' if x > mediana_calidad else 'Baja'
)

print(f"üîß GRUPOS DE PROCESOS ADMINISTRATIVOS (n=18):")
grupos_proc = df_procesos['GRUPO_PROCESOS'].value_counts()
for grupo, count in grupos_proc.items():
    print(f"   {grupo}: {count} participantes ({count/18*100:.1f}%)")

print(f"")
print(f"üìä GRUPOS DE CALIDAD DE SERVICIO (n=70):")
grupos_cal = df_calidad['GRUPO_CALIDAD'].value_counts()
for grupo, count in grupos_cal.items():
    print(f"   {grupo}: {count} participantes ({count/70*100:.1f}%)")

# An√°lisis descriptivo por grupos - Procesos
print(f"\nüìà AN√ÅLISIS POR GRUPOS - PROCESOS:")
print("-"*40)
for grupo in ['Bajos', 'Altos']:
    datos_grupo = df_procesos[df_procesos['GRUPO_PROCESOS'] == grupo]
    if len(datos_grupo) > 0:
        media = datos_grupo['PROCESOS_GLOBAL'].mean()
        std = datos_grupo['PROCESOS_GLOBAL'].std()
        print(f"   {grupo}: M = {media:.2f} ¬± {std:.2f} (n = {len(datos_grupo)})")

# An√°lisis descriptivo por grupos - Calidad
print(f"\nüìà AN√ÅLISIS POR GRUPOS - CALIDAD:")
print("-"*40)
for grupo in ['Baja', 'Alta']:
    datos_grupo = df_calidad[df_calidad['GRUPO_CALIDAD'] == grupo]
    if len(datos_grupo) > 0:
        media = datos_grupo['CALIDAD_GLOBAL'].mean()
        std = datos_grupo['CALIDAD_GLOBAL'].std()
        print(f"   {grupo}: M = {media:.2f} ¬± {std:.2f} (n = {len(datos_grupo)})")

# Visualizaciones adicionales
plt.subplot(2, 2, 3)
# Boxplot por grupos de calidad
df_calidad.boxplot(column='CALIDAD_GLOBAL', by='GRUPO_CALIDAD', ax=plt.gca())
plt.title('Distribuci√≥n de Calidad por Grupos')
plt.suptitle('')  # Remover t√≠tulo autom√°tico
plt.ylabel('Calidad Global')

plt.subplot(2, 2, 4)
# Histograma de correlaci√≥n
plt.hist(residuos, bins=8, alpha=0.7, color='lightblue', edgecolor='black')
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Residuos')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# TABLA RESUMEN DE RESULTADOS
print(f"\nüìã TABLA RESUMEN DE RESULTADOS PRINCIPALES")
print("="*60)

resultados_principales = pd.DataFrame({
    'Estad√≠stico': [
        'Muestra Calidad (n)',
        'Muestra Procesos (n)',
        'Media Calidad Global',
        'Media Procesos Global',
        'Correlaci√≥n (r)',
        'R¬≤ Regresi√≥n Simple',
        'R¬≤ Regresi√≥n M√∫ltiple',
        'p-valor Correlaci√≥n',
        'Significancia (Œ±=0.05)',
        'Alpha Cronbach Calidad',
        'Alpha Cronbach Procesos'
    ],
    'Valor': [
        f"{len(df_calidad)}",
        f"{len(df_procesos)}",
        f"{df_calidad['CALIDAD_GLOBAL'].mean():.3f}",
        f"{df_procesos['PROCESOS_GLOBAL'].mean():.3f}",
        f"{stat_corr:.3f}",
        f"{r2:.3f}",
        f"{r2_multiple:.3f}",
        f"{p_corr:.3f}",
        conclusion,
        f"{alpha_total_calidad:.3f}",
        f"{alpha_total_procesos:.3f}"
    ]
})

print(resultados_principales.to_string(index=False))

# CONCLUSIONES Y RECOMENDACIONES
print(f"\nüéØ CONCLUSIONES PRINCIPALES")
print("="*50)

print(f"1. RELACI√ìN ENTRE VARIABLES:")
if abs(stat_corr) >= 0.5:
    print(f"   ‚úÖ Existe una relaci√≥n FUERTE entre procesos y calidad (r = {stat_corr:.3f})")
elif abs(stat_corr) >= 0.3:
    print(f"   ‚úÖ Existe una relaci√≥n MODERADA entre procesos y calidad (r = {stat_corr:.3f})")
else:
    print(f"   ‚ö†Ô∏è La relaci√≥n entre procesos y calidad es D√âBIL (r = {stat_corr:.3f})")

print(f"")
print(f"2. SIGNIFICANCIA ESTAD√çSTICA:")
if p_corr < 0.05:
    print(f"   ‚úÖ La relaci√≥n es estad√≠sticamente significativa (p = {p_corr:.3f})")
    print(f"   üìä Se puede afirmar que los procesos influyen en la calidad")
else:
    print(f"   ‚ö†Ô∏è La relaci√≥n NO es estad√≠sticamente significativa (p = {p_corr:.3f})")
    print(f"   üìä No se puede confirmar influencia estad√≠stica")

print(f"")
print(f"3. CAPACIDAD PREDICTIVA:")
if r2 >= 0.5:
    print(f"   üéØ ALTA: El modelo explica {r2*100:.1f}% de la variabilidad")
elif r2 >= 0.25:
    print(f"   üéØ MODERADA: El modelo explica {r2*100:.1f}% de la variabilidad")
else:
    print(f"   üéØ BAJA: El modelo solo explica {r2*100:.1f}% de la variabilidad")

print(f"")
print(f"4. CONFIABILIDAD DE INSTRUMENTOS:")
if alpha_total_calidad >= 0.8 and alpha_total_procesos >= 0.8:
    print(f"   ‚úÖ Ambos instrumentos tienen BUENA confiabilidad")
elif alpha_total_calidad >= 0.7 and alpha_total_procesos >= 0.7:
    print(f"   ‚úÖ Ambos instrumentos tienen confiabilidad ACEPTABLE")
else:
    print(f"   ‚ö†Ô∏è Uno o ambos instrumentos requieren REVISI√ìN")

print(f"\nüí° RECOMENDACIONES PARA LA INSTITUCI√ìN:")
print("-"*50)

# Identificar dimensiones m√°s y menos problem√°ticas
dimension_menor = min(medias_calidad, key=lambda x: x[1])
dimension_mayor = max(medias_calidad, key=lambda x: x[1])

print(f"üî¥ PRIORIDAD ALTA - Mejorar: {dimension_menor[0]} (Media: {dimension_menor[1]:.2f})")
print(f"üü¢ FORTALEZA - Mantener: {dimension_mayor[0]} (Media: {dimension_mayor[1]:.2f})")

if stat_corr > 0 and p_corr < 0.05:
    print(f"üìà ENFOQUE: Invertir en mejora de procesos administrativos")
    print(f"   - Cada mejora en procesos se asocia con mejora en calidad")
    print(f"   - Priorizar las dimensiones de procesos con mayor impacto")

print(f"\n‚úÖ AN√ÅLISIS DE REGRESI√ìN Y PRUEBAS DE HIP√ìTESIS COMPLETADO")
print("üìå SIGUIENTE PASO: Ejecutar Parte 7 - Resumen final y exportaci√≥n")
```
